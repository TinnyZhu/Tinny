```python
# (2)

xvalue=np.linspace(0,5,10000)
terms=np.array([1,2,3,10])
yvalue=[[],[],[],[]]
for i in range(0,4):
    value=[]
    for j in xvalue:
        summ=0
        for k in range(0,terms[i]+1):
            summ=summ+j**k/np.math.factorial(k)
        value=np.append(value,summ)
    yvalue[i]=value
    plt.plot(xvalue,yvalue[i])
yvaluet=np.exp(xvalue)
plt.plot(xvalue,yvaluet)
```
